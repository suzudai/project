```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.nankankeiba.com/uma_info/2013103166.do'
dfs = pd.read_html(url)
```

```python
# feature = ['date', 'place','length','weather_state','popularity','time','weight','load_wei','money']
df = pd.DataFrame(dfs[4])
df1 = df.drop(['R','レース名', '馬番', '差/事故', '上3F', 'コーナー通過順', '騎手', '調教師'], axis = 1)
```

```python
df1.rename(columns={'年月日': 'date', '競馬場':'place','距離':'length','人気':'popularity',
                    'タイム': 'time','体重': 'weight', '負担重量': 'load_wei','獲得賞金（円）': 'money', '着順': 'order', '0':'weather'})
```

```python
df2 = df1['天候馬場'].str.split("/", expand=True)
df3 = pd.concat([df1, df1['天候馬場'].str.split("/", expand=True)], axis = 1).drop('天候馬場', axis = 1)
df3
```

```python
from sklearn.model_selection import train_test_split
keiba_train, keiba_test = train_test_split(df3, test_size=0.2)
print(keiba_train)
```
